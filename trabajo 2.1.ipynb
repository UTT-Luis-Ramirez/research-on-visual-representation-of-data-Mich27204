## Historigrama

#### Como se crean?
Paso 1: Recolecta y organiza tus datos:
Primero, necesitas un conjunto de datos que deseas analizar. Por ejemplo, podrías tener una lista de edades de un grupo de personas.

Paso 2: Divide los datos en intervalos (bins):
Decide cuántos intervalos (bins) deseas utilizar. Los intervalos representan los rangos de valores en los que los datos se agruparán. Por ejemplo, si estás trabajando con edades, podrías tener intervalos como 0-10, 11-20, 21-30, etc.

Paso 3: Cuenta los datos en cada intervalo:
Cuenta cuántos datos caen dentro de cada intervalo. Esto se llama la frecuencia de cada intervalo.

Paso 4: Crea el gráfico:
Dibuja un gráfico de barras donde el eje x representa los intervalos y el eje y representa la frecuencia de cada intervalo. Cada barra del histograma representa un intervalo y su altura corresponde a la frecuencia de ese intervalo.

#### Para que son?
##### Visualizacion de datos:
permiten visualizar la distribucion de un conjunto de datos de manera clara y concisa. facilita la comprension de como se distribuyen los datos y cuales son los valores mas comunes.
##### Analisis Estadistico:
Son útiles para realizar análisis estadísticos. Con ellos, puedes identificar patrones, tendencias y anomalías en tus datos. Por ejemplo, puedes ver si tus datos siguen una distribución normal, detectar sesgos o entender la variabilidad.
##### Toma de decisiones:
En entornos empresariales, los histogramas pueden ayudar a tomar decisiones informadas. Por ejemplo, en el control de calidad, se pueden usar histogramas para identificar defectos en productos y mejorar los procesos de producción.
##### Educacion: 
En el ámbito educativo, los histogramas son una excelente herramienta para enseñar conceptos estadísticos y matemáticos. Ayudan a los estudiantes a visualizar y comprender la teoría detrás de las distribuciones de datos.

#### En que forma de contexto se utiliza?
Los histogramas se utilizan en una variedad de contextos y campos para analizar la distribución de los datos y extraer información útil. Aquí te menciono algunos contextos en los que se utilizan frecuentemente:

Contextos Académicos y Educativos.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
data=[22,25,23,29,31,20,22,24,28,30,33,35,21,26]

# Crear un histograma
plt.hist(data, bins=5, color='blue', edgecolor='black')

# Añadir título y etiquetas
plt.title('Histograma de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar grafico
plt.show()
```

## Diagrama de Tallo y hoja:

#### Como se crean:
se crea recolectando datos y organizandolos en orden ascendente.

Divide los datos en tallos y hojas donde el tallo representa los primeros digitos de los numeros y la hoja representa el ultimo.

se crea el diagrama en una linea vertical donde los tallos ala izquierda de la linea y las hojas ala derecha,agrupa los datos correspondiente a cada tallo y hoja.

#### Para que son:
son utiles para una visualizacion de datos que permiten ver la distribucion de los datos de una forma mas clara.

Tambien ayuda a detectar errores(Outliers).

Facilita la comparacion visual de entre dos conjuntos de datos.

Son herramientas valiosas para estudiar.

Proporcionan un resumen visual detallado sin perder informacion importante.

#### Como son?
Tallo: El número a la izquierda de la línea vertical representa el primer dígito de cada número en tus datos (en este caso, las decenas).

Hoja: Los números a la derecha de la línea vertical representan el segundo dígito de cada número (las unidades).

#### En que forma de contexto se utiliza?
Contextos Académicos y Educativos.

Estudios estadísticos: Ayudan a enseñar conceptos de distribución de datos y a realizar análisis descriptivos.

Investigaciones científicas: Útiles para visualizar datos experimentales y observar patrones.

Un ejemplo coridiano seria:crear un diagrama de tallo y hoja para los habito de estudio y ver como distribuir nuestro tiempo.

Ejercicio fisico: registrar la fecuencia y duracion de tus entrenamientos para ajustar rutinas.

```python
import numpy as np

# Datos de ejemplo
data = np.array([23, 26, 28, 34, 89, 41, 45, 46, 49, 52])

# Crear tallos y hojas
stems, leaves = np.divmod(data, 10)

# Organizar datos en un diccionario
stem_leaf = {}
for stem, leaf in zip(stems, leaves):
    if stem not in stem_leaf:
        stem_leaf[stem] = []
    stem_leaf[stem].append(leaf)

# Imprimir diagrama de tallo y hoja
for stem in sorted(stem_leaf):
    print(f'{stem} | {" ".join(map(str, sorted(stem_leaf[stem])))}')
```

## Diagram de caja
Recolecta y organiza los datos: Reúne el conjunto de datos que deseas analizar.

Calcula los valores estadísticos clave: Mínimo, primer cuartil (Q1), mediana (Q2), tercer cuartil (Q3) y máximo.

Dibuja la caja: La caja se extiende desde Q1 hasta Q3, con una línea en la mediana (Q2).

Dibuja los bigotes: Los bigotes se extienden desde Q1 hasta el mínimo y desde Q3 hasta el máximo. Los outliers se representan como puntos fuera de los bigotes.

#### Para que son?
Visualizar la distribución: Muestran la dispersión y centralidad de los datos.

Identificar outliers: Detectan valores atípicos.

Comparar conjuntos de datos: Facilitan la comparación visual de varios conjuntos.

Resumir datos: Proporcionan un resumen visual con mediana, cuartiles y extremos.

Aplicaciones en diferentes campos: Investigaciones científicas, negocios, salud, educación, entre otros.

#### Como son?

 23   26   28   34   38   41   45   46   49   52

 +---------+--------+-------+--------+---------+
|  Min/Q1 | Median | Q3/Max| Outliers|
+---------+--------+-------+--------+---------+

Componentes del Diagrama de Caja:
Caja: La parte central que va desde el primer cuartil (Q1) hasta el tercer cuartil (Q3).

Línea dentro de la caja: Representa la mediana (Q2) del conjunto de datos.

Bigotes: Se extienden desde los extremos de la caja (Q1 y Q3) hasta los valores mínimo y máximo (excluyendo outliers).

Outliers: Valores atípicos que se muestran como puntos individuales fuera de los bigotes.

#### Para que sirven?
Visualización de la Distribución de Datos: Muestran cómo se dispersan los datos y su centralidad.

Identificación de Valores Atípicos (Outliers): Detectan valores que se desvían significativamente del resto.

Comparación de Conjuntos de Datos: Facilitan la comparación visual de diferentes conjuntos de datos.

Resumen de Datos Estadísticos: Proporcionan información clave como la mediana, los cuartiles y los valores extremos.

Aplicaciones en Diferentes Campos: Se utilizan en investigaciones científicas, negocios, salud y educación.

#### En que forma de contexto se utiliza?
Contextos Académicos y Educativos
Estudios estadísticos: Ayudan a enseñar conceptos de distribución de datos y realizar análisis descriptivos.

Investigaciones científicas: Visualizan datos experimentales para observar patrones y comparar resultados de diferentes estudios.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
data = [23, 26, 28, 34, 38, 41, 45, 46, 49, 52]

# Crear diagrama de caja
plt.boxplot(data)

# Añadir títulos y etiquetas
plt.title('Diagrama de Caja')
plt.xlabel('Grupo')
plt.ylabel('Edad')

# Mostrar gráfico
plt.show()
```
